## FEATURE ENGINEERING

```python
# lets create the function that extracts date features -> day of the week, week, month, quarter, year
```

```python
def extract_features(dataframe):
    """
    this function extracts date features from a date index, saving it to a new column
    """
    dataframe=dataframe.copy()
    dataframe["day of week"] = dataframe.index.dayofweek 
    dataframe["quarter"] = dataframe.index.quarter
    dataframe["month"] = dataframe.index.month
    dataframe["year"] = dataframe.index.year
    dataframe["week"] = dataframe.index.isocalendar().week.astype(int)

    return dataframe
```

## CREATING SEVERAL FORECAST MODELS AND EVALUATING

FOR EACH MODEL CANDIDATE

for every product 
1. We create a model 

2. We evaluate the model with mape, mae, and rmse scores

3. Visualize the forecast on the test set

```python
##Creating a function that would help evaluate the models

import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error


def compute_scores(predicted_values, actual_values):
  # Calculate mae
    mae = mean_absolute_error(actual_values, predicted_values)
  # Calculate rmse
    rmse = np.sqrt(mean_squared_error (actual_values,predicted_values))
  # Calculate mape
    mape = np.mean(abs(actual_values-predicted_values)/(actual_values + 1))* 100\

    return mae, rmse, mape
```

```python
# features, target_variable 

features = ["day of week", "quarter", "month", "year",'week']

target = ["Demand"]
```

```python
from xgboost import XGBRegressor 
# number of estimators = 1000
# learning rate = 0.001
```

```python
#for each product 
#subselect the product data
#collect features and target
#train model
# evaluate model
# stores results 
```

```python
xgboost_model_results = {'product' : [], 'mae' : [], 'rmse' : [] ,'mape' : [] , 'pred' : [], 'y_test' : []}
for prod_id in df_train["Product_id"].unique():
    df_train_prod_id = df_train[df_train['Product_id'] == prod_id].resample ("M").sum()
    df_test_prod_id = df_test[df_test['Product_id'] == prod_id].resample ("M").sum()

    #extract features
    X_train = df_train_prod_id [features]
    y_train = df_train_prod_id [target]
    X_test =  df_test_prod_id [features]
    y_test =  df_test_prod_id [target]

    # Initialize and train model
    model = XGBRegressor(n_estimators=1000,early_stopping_rounds= 50,learning_rate = 0.01)
    model.fit(X_train, y_train, eval_set = [(X_train,y_train),(X_test, y_test)], verbose = 50)

    #evaluating and storing results 
    pred= model.predict(X_test)
    mae,rmse, mape = compute_scores(pred, y_test.values)

    xgboost_model_results['product'].append(prod_id)
    xgboost_model_results['mae'].append(mae)
    xgboost_model_results['rmse'].append(rmse)
    xgboost_model_results['mape'].append(mape)
    xgboost_model_results['pred'].append(pred)
    xgboost_model_results['y_test'].append(y_test.Demand.values)

x_gboost_model_results_monthly = pd.DataFrame(xgboost_model_results)
```

```python
#imports

#adjusts

from statsmodels.tsa.arima.model import ARIMA

arima_model_results = {'product' : [], 'mae' : [], 'rmse' : [] ,'mape' : [] , 'pred' : [], 'y_test' : []}

for prod_id in df_train["Product_id"].unique():
    df_train_prod_id = df_train[df_train['Product_id'] == prod_id].resample ("W").sum()
    df_test_prod_id = df_test[df_test['Product_id'] == prod_id].resample ("W").sum()

    #extract features
    y_train = df_train_prod_id [target]
    y_test =  df_test_prod_id [target]

    # Initialize and train model
    order = (1,1,1)#p(autoregressive) d(differencing component) q(moving average)
    model= ARIMA(y_train, order = order)
    model_fit=model.fit()

    #evaluating and storing results 
    pred= model_fit.forecast(len(y_test))
    y_test=y_test.values.reshape(-1)
    mae,rmse, mape = compute_scores(pred,y_test)

    arima_model_results['product'].append(prod_id)
    arima_model_results['mae'].append(mae)
    arima_model_results['rmse'].append(rmse)
    arima_model_results['mape'].append(mape)
    arima_model_results['pred'].append(pred.values)
    arima_model_results['y_test'].append(y_test)

arima_model_results_weekly = pd.DataFrame(arima_model_results)
```

## SARIMAX MODEL

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarimax_model_results = {'product' : [], 'mae' : [], 'rmse' : [] ,'mape' : [] , 'pred' : [], 'y_test' : [] }

for prod_id in df_train["Product_id"].unique():
    df_train_prod_id = df_train[df_train['Product_id'] == prod_id].resample ("W").sum()
    df_test_prod_id = df_test[df_test['Product_id'] == prod_id].resample ("W").sum()

    #extract features
    y_train = df_train_prod_id [target]
    y_test =  df_test_prod_id [target]

    # Initialize and train model
    order = (1,1,1)#p(autoregressive) d(differencing component) q(moving average)
    seasonal_order = (1, 1, 1, 52)  # P, D, Q, m (seasonality component)

    model= SARIMAX(y_train, order = order, seasonal_order=seasonal_order)
    model_fit=model.fit()

    #evaluating and storing results 
    pred= model_fit.forecast(len(y_test))
    y_test=y_test.values.reshape(-1)
    mae,rmse, mape = compute_scores(pred, y_test)

    sarimax_model_results['product'].append(prod_id)
    sarimax_model_results['mae'].append(mae)
    sarimax_model_results['rmse'].append(rmse)
    sarimax_model_results['mape'].append(mape)
    sarimax_model_results['pred'].append(pred.values)
    sarimax_model_results['y_test'].append(y_test)

sarimax_model_results_weekly = pd.DataFrame(sarimax_model_results)
```

```python
## Exponential Smoothing Model

 # trend: ADDITIVE or multiplicative
 # seasonality: ADDITIVE or multiplicative
 # seasonal_periods = 52 
 
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

exp_model_results = {'product' : [], 'mae' : [], 'rmse' : [] ,'mape' : [] , 'pred' : [], 'y_test' : [] }

for prod_id in df_train["Product_id"].unique():
    df_train_prod_id = df_train[df_train['Product_id'] == prod_id].resample ("W").sum()
    df_test_prod_id = df_test[df_test['Product_id'] == prod_id].resample ("W").sum()

    #extract features
    y_train = df_train_prod_id [target]
    y_test =  df_test_prod_id [target]

    # Initialize and train model
    trend ="add"
    seasonal= "add"
    seasonal_periods = 52
    model= ExponentialSmoothing (y_train,seasonal=seasonal,seasonal_periods=seasonal_periods,trend=trend)
    model_fit=model.fit()
    #evaluating and storing results 
    pred= model_fit.forecast(steps= len(y_test))
    y_test=y_test.values.reshape(-1)
    mae,rmse, mape = compute_scores(pred, y_test)

    exp_model_results['product'].append(prod_id)
    exp_model_results['mae'].append(mae)
    exp_model_results['rmse'].append(rmse)
    exp_model_results['mape'].append(mape)
    exp_model_results['pred'].append(pred.values)
    exp_model_results['y_test'].append(y_test)

exp_model_results_weekly = pd.DataFrame(exp_model_results)
```

1. Sarima and exponential smoothing are better due to the seasonality components.

3. For all the four models try tweaking training hyperparameters (tuning) to get better performance 

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

product_forecast_models = []
# Create dictionary for the results of the Exponential Smoothing model
exp_smoothing_model_results = {"product": [], "mae": [], "rmse": [], "mape": [], "pred": [], "y_test": []}

# For each product 
for prod_id in df_train["Product_id"].unique():
    # Subselect the product data 
    df_train_prod_id = df_train[df_train["Product_id"] == prod_id].resample("W").sum()
    df_test_prod_id = df_test[df_test["Product_id"] == prod_id].resample("W").sum()

    # Extract features and target variables
    y_train = df_train_prod_id[target]
    y_test = df_test_prod_id[target]

    # Initialize and train Exponential Smoothing model
    model = ExponentialSmoothing(y_train, trend='add', seasonal='add', seasonal_periods=52)
    model_fit = model.fit()


    # Evaluating and storing results
    pred = model_fit.forecast(len(y_test))
    y_test = y_test.values.reshape(-1)
    product_forecast_models.append(model_fit)

    mae, rmse, mape = compute_scores(pred, y_test)
    exp_smoothing_model_results["product"].append(prod_id)
    exp_smoothing_model_results["mae"].append(mae)
    exp_smoothing_model_results["rmse"].append(rmse)
    exp_smoothing_model_results["mape"].append(mape)
    exp_smoothing_model_results["pred"].append(pred.values)
    exp_smoothing_model_results["y_test"].append(y_test)

# Convert results to DataFrame
exp_smoothing_model_results_weekly = pd.DataFrame(exp_smoothing_model_results)
```

```python
def forecast(weeks,product= None):
    """
    weeks ==> int 
    product ==> int or all models 
    this function receives the number of weeks to forecast on 
    and what model to use 
    """
    if product:
       models =[product_forecast_models[product]]
    else:
       models = product_forecast_models
    forecast_values = [model.forecast(steps = weeks) for model in models]
    return forecast_values    
```
